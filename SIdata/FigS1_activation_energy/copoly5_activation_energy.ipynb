```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import glob
import json
from matplotlib.patches import Patch
```

## Standard Averaging

```python
results_Ea_all = []
stdev_Ea_all = []
events_all = []
#Results kept in nested lists
#First index in lists - chain stiffness 0=flex 1= stiff

#Second index is E_barr: 
#0 -  E_barr AA,BB = 1 kT E_barr AB = 3kT
#1 -  E_barr AA,BB = 3 kT E_barr AB = 3kT
#3 -  E_barr AA,BB = 3 kT E_barr AB = 1kT

#Third index is bond event sequence pair
#0 - AA , 1 - AB , 2 - BB
events = np.array([0,0])
cnt = 0
for barrier_path in ['./Flex/','./Stiff/']:
    results_Ea = []
    stdev_Ea = []
    for bar_height in ['*3.0rbAA_3.0rbBB_9.0rbAB*_v5.json','*9.0rbAA_9.0rbBB_9.0rbAB*_v5.json','*9.0rbAA_9.0rbBB_3.0rbAB*_v5.json']:
        barfiles = sorted(glob.glob(barrier_path+bar_height))

        barAA_list = []
        barAB_list = []
        barBB_list = []

        for cntr, barfile in enumerate(barfiles):
            with open(barfile,'r') as f:
                record = json.load(f)
            barAA_list.append(np.array([i['energy']['total'] for i in record if i['type']==[2,2]]))
            barAB_list.append(np.array([i['energy']['total'] for i in record if i['type']==[2,3] or i['type']==[3,2]]))
            barBB_list.append(np.array([i['energy']['total'] for i in record if i['type']==[3,3]]))



        barAA=[val/3 for i in barAA_list for val in i]
        barAB=[val/3 for i in barAB_list for val in i]  
        barBB=[val/3 for i in barBB_list for val in i]
        events[cnt] += len(barAA)+len(barAB)+len(barBB)
        for entry in barAA:
            events_all.append(entry)
        for entry in barAB:
            events_all.append(entry)
        for entry in barBB:
            events_all.append(entry)


        barAA_avg = np.mean(np.array(barAA))
        barAA_std = np.std(np.array(barAA))
        barAB_avg = np.mean(np.array(barAB))
        barAB_std = np.std(np.array(barAB))
        barBB_avg = np.mean(np.array(barBB))
        barBB_std = np.std(np.array(barBB))

        results_Ea.append([barAA_avg,barAB_avg,barBB_avg])
        stdev_Ea.append([barAA_std,barAB_std,barBB_std])
    results_Ea_all.append(results_Ea)
    stdev_Ea_all.append(stdev_Ea)
    cnt+=1
```

```python
font = {'size':20}
matplotlib.rc('axes',labelsize=28)
matplotlib.rc('font',**font)
matplotlib.rcParams['figure.dpi'] = 300
matplotlib.rcParams.update({'errorbar.capsize': 10})

titles = ['Flexible Chains: $l_p = 3.5$','Stiff Chains: $l_p = 16.5$']

fig, ax = plt.subplots(1,2,figsize= (24,8))

for i in [0,1]:
    ax[i].bar(0-(0.5/3),results_Ea_all[i][0][0]+1,width=0.5/3,yerr=stdev_Ea_all[i][0][0],color='r',alpha=0.5)
    ax[i].bar(0,results_Ea_all[i][0][1]+3,width=0.5/3, yerr=stdev_Ea_all[i][0][1],color='purple',alpha=0.5)
    ax[i].bar(0+(0.5/3),results_Ea_all[i][0][2]+1,width=0.5/3,yerr=stdev_Ea_all[i][0][2], color='b',alpha=0.5)
    
    ax[i].bar(1-(0.5/3),results_Ea_all[i][1][0]+3,width=0.5/3,yerr=stdev_Ea_all[i][1][0],color='r',alpha=0.5)
    ax[i].bar(1,results_Ea_all[i][1][1]+3,width=0.5/3, yerr=stdev_Ea_all[i][1][1],color='purple',alpha=0.5)
    ax[i].bar(1+(0.5/3),results_Ea_all[i][1][2]+3,width=0.5/3,yerr=stdev_Ea_all[i][1][2], color='b',alpha=0.5)
    
    ax[i].bar(2-(0.5/3),results_Ea_all[i][2][0]+3,width=0.5/3,yerr=stdev_Ea_all[i][2][0],color='r',alpha=0.5)
    ax[i].bar(2,results_Ea_all[i][2][1]+1,width=0.5/3, yerr=stdev_Ea_all[i][2][1],color='purple',alpha=0.5)
    ax[i].bar(2+(0.5/3),results_Ea_all[i][2][2]+3,width=0.5/3,yerr=stdev_Ea_all[i][2][2], color='b',alpha=0.5)    
    
    
    ax[i].bar(0-(0.5/3),results_Ea_all[i][0][0],width=0.5/3, color='r',alpha=0.5)
    ax[i].bar(0,results_Ea_all[i][0][1],width=0.5/3, color='purple',alpha=0.5)
    ax[i].bar(0+(0.5/3),results_Ea_all[i][0][2],width=0.5/3, color='b',alpha=0.5)
    
    ax[i].bar(1-(0.5/3),results_Ea_all[i][1][0],width=0.5/3,color='r',alpha=0.5)
    ax[i].bar(1,results_Ea_all[i][1][1],width=0.5/3,color='purple',alpha=0.5)
    ax[i].bar(1+(0.5/3),results_Ea_all[i][1][2],width=0.5/3, color='b',alpha=0.5)
    
    ax[i].bar(2-(0.5/3),results_Ea_all[i][2][0],width=0.5/3,color='r',alpha=0.5)
    ax[i].bar(2,results_Ea_all[i][2][1],width=0.5/3,color='purple',alpha=0.5)
    ax[i].bar(2+(0.5/3),results_Ea_all[i][2][2],width=0.5/3, color='b',alpha=0.5)    
    
    
    
    label1 = r'$E_{barr}^{AA,BB} = 1.00 k_{B}T$'+'\n'+'$E_{barr}^{AB} = 3.00 k_{B}T$'
    label2 = r'$E_{barr}^{AA,BB} = 3.00 k_{B}T$'+'\n'+'$E_{barr}^{AB} = 3.00 k_{B}T$'
    label3 = r'$E_{barr}^{AA,BB} = 3.00 k_{B}T$'+'\n'+'$E_{barr}^{AB} = 1.00 k_{B}T$'
    
    ax[i].set_xticks([0,1,2])
    ax[i].set_xticklabels([label1,label2,label3])
    ax[i].set_ylim((0,15));
    ax[i].set_title(titles[i])
    
    
    if i==0:
        legend_elements = [Patch(facecolor='r', edgecolor='r',label='AA'),Patch(facecolor='purple', edgecolor='purple',label='AB'),Patch(facecolor='b', edgecolor='b',label='BB')]
        ax[i].legend(handles=legend_elements, loc='upper center', ncol=3)
        ax[i].set_ylabel(r'$E_a \ (k_BT)$')

print('Flexible events:')
print(events[0])        
print('Stiff events:')
print(events[1])


fig.subplots_adjust(wspace=0.1)
plt.show()
fig.savefig('copoly5_FigSI_ActivationEnergy.png',dpi=300, bbox_inches='tight')
```
