```python
import numpy as np
from scipy import integrate
from scipy import stats
import matplotlib.pylab as plt
import matplotlib
from sklearn.metrics import r2_score
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

## Arrhenius Pre-factor from Fit to Simulation Results

```python
#epsilon values from simulations
eps_array2 = np.array([0,0.016666667,0.033333333,0.05,0.066666667,0.09999999,0.15,0.2,0.25])


#Arrhenius pre-exponential A factors obtained from simulation at each epsilon
y_Nhu = np.array([0.6151745942359972,0.6333451450392329,0.6384711790706146,0.6363153322843144,0.6558098101358059,0.6812190499375526,0.697453078868466,0.7483697879530885,0.7753650909253442])


#St Dev vals in A from Nhu
y_std = np.array([0.00544321671179016,0.00978287717868724,0.006591302302070473,0.007274848504310288,0.007388299622405438,0.010420641234491565,0.008139359473214812,0.009420500089548612,0.0097389])






#Obtaining rho_coll at the epsilon values used in simulation
rho_coll_exp = []
for eps in eps_array2:
    #integrated g(r)
    I1 = integrate.quad(lambda r: ((r**2)*np.exp(-4*eps*(r**(-12)-r**(-6)))),int_start,int_end)
    #volume of spherical shell integrated
    I2 = integrate.quad(lambda r: r**2,int_start,int_end)
    
    
    #integrated g(r) - repulsive
    I3 = integrate.quad(lambda r: ((r**2)*np.exp(-(4*eps_rep*(r**(-12)-r**(-6))+(eps_rep-eps)))),int_rep,int_start)
    #volume of spherical shell integrated - repulsive
    I4 = integrate.quad(lambda r: r**2,int_rep,int_start)
    
    
    if full_range:
        rho_coll_exp.append(I1[0]+I3[0])
    else:
        rho_coll_exp.append(I1[0])

rho_coll_exp = np.array(rho_coll_exp)
```

## Integrating RDF

```python
#large range of LJ epsilon values
eps_array = np.linspace(0.0,1.0,251)
# rho_coll is the collisional density correction factor (multiply by bulk density to get effective collisional density at a given epsilon value)
# taken over a broad array of epsilon values for testing purposes
rho_coll = []

#useful constants
int_start = 2**(1/6)
int_rep = 0
int_end = 2.5
eps_rep = 4.0/3.0
vol_full = 4/3*np.pi*(2.5)**3
rho = (7200/(50**3)/2)
spherical_int_att = []
spherical_int_full = []

#Integrate over the whole LJ range or only attractive portion
full_range = True

for eps in eps_array:
    #integrated g(r)
    I1 = integrate.quad(lambda r: ((r**2)*np.exp(-4*eps*(r**(-12)-r**(-6)))),int_start,int_end)
    #volume of spherical shell integrated
    I2 = integrate.quad(lambda r: r**2,int_start,int_end)
    
    
    #integrated g(r) - repulsive
    I3 = integrate.quad(lambda r: ((r**2)*np.exp(-(4*eps_rep*(r**(-12)-r**(-6))+(eps_rep-eps)))),int_rep,int_start)
    #volume of spherical shell integrated - repulsive
    I4 = integrate.quad(lambda r: r**2,int_rep,int_start)
    
    if full_range:
        rho_coll.append((I1[0]+I3[0]))
    else:
        rho_coll.append(I1[0])
    
    
    spherical_int_full.append(I1[0]+I3[0])
    spherical_int_att.append(I1[0])
    
rho_coll = np.array(rho_coll)
spherical_int_att = np.array(spherical_int_att)
spherical_int_full = np.array(spherical_int_full)
kT_index = np.where(eps_array==1)[0][0]
```

## Linear Mapping - 
## Measured $A_{ij}(\epsilon_{ij})$ to Theoretical $\rho_{\rm coll}(\epsilon_{ij})$

```python
# Regression
linfit_exp = np.polyfit(rho_coll_exp,y_Nhu,1)
linfit_exp_line = linfit_exp[0]*rho_coll_exp + linfit_exp[1]
lin_exp_r2 = r2_score(linfit_exp_line,y_Nhu)

#Plot params (global)
smallsize = 14
largesize = 18
font = {'family' : 'DejaVu Sans',
        'weight' : 'normal',
        'size'   : largesize}
matplotlib.rc('font',**font)
matplotlib.rc('axes',labelsize=largesize)

#Plotting
fig, host = plt.subplots(1,1)
host.tick_params(axis='both', which='major', labelsize=largesize)
host.set_ylabel(r'$A_{ij}(\epsilon_{ij}$)')
host.set_xlabel(r'$\sigma_{ij}(\epsilon_{ij})$')
plt.errorbar(rho_coll_exp,y_Nhu, yerr=y_std, fmt='o', capsize=2.)
host.plot(rho_coll_exp,linfit_exp_line,'--k',label='Linear Fit')
host.text(0.07,0.6,r'$y=mx+b$'+'\n'+'m={:.2f}'.format(linfit_exp[0])+'\n'+'b={:.2f}'.format(linfit_exp[1])+'\n'+'$R^2=${:.3f}'.format(lin_exp_r2),transform = host.transAxes,color='k');

#Fitting error
#https://stats.stackexchange.com/questions/175026/formula-for-95-confidence-interval-for-r2
#https://books.google.com/books?id=gkalyqTMXNEC&pg=PA88
SE_r2 = ((4*lin_exp_r2**2*(1-lin_exp_r2**2)**2*(len(eps_array2)-2)**2)/((len(eps_array2)**2-1)*(len(eps_array2)+3)))**0.5
#print(SE_r2)



#Extrapolation of trend
A_final = linfit_exp[0]*rho_coll[kT_index] + linfit_exp[1]
A_0 = linfit_exp[0]*rho_coll[0] + linfit_exp[1]
print(r"At $\epsilon_{ij}$ = "+'{:.2f}'.format(eps_array[kT_index])+r" $k_{\rm B}T$")
#print(r"Theoretical value of $\rho_{\rm coll}$ = "+'{:.4f}'.format(rho_coll[kT_index]))
print(r"Predicted value of $A_{ij}$ = "+'{:.4f}'.format(A_final))
print(r"At $\epsilon_{ij}$ = "+'{:.2f}'.format(eps_array[0])+r" $k_{\rm B}T$")
#print(r"Theoretical value of $\rho_{\rm coll}$ = "+'{:.4f}'.format(rho_coll[0]))
print(r"Predicted value of $A_{ij}$ = "+'{:.4f}'.format(A_0))
print(r"Predicted value of $A$ ratio = "+'{:.4f}'.format(A_final/A_0))



```

```python
fig, host = plt.subplots(1,1,figsize = (9,6))



host.set_xlim((-0.025,1.025))
host.set_xlabel(r'$\epsilon_{ij} \ (k_{\rm B}T)$')
host.set_ylabel(r'$A_{ij}(\epsilon_{ij}$)')
plt.errorbar(eps_array2,y_Nhu, yerr=y_std, fmt='o', capsize=2., color='k', label='Simulation Data')
host.plot(eps_array,linfit_exp[0]*np.array(rho_coll) + linfit_exp[1],label='Numerical Fit to Theory',color='blue')
host.fill_between(eps_array,linfit_exp[0]*np.array(rho_coll) + linfit_exp[1] - SE_r2, linfit_exp[0]*np.array(rho_coll) + linfit_exp[1] + SE_r2, color='b', alpha=.1)
host.legend(loc='lower right');



axins = inset_axes(host, width='40%',height='35%', loc='upper left', borderpad=1.)
axins.set_ylabel(r'$A_{ij}(\epsilon_{ij})$',fontsize = smallsize)
axins.set_xlabel(r'$\sigma_{ij}(\epsilon_{ij})$',fontsize = smallsize)
axins.tick_params(axis='both', which='major', labelsize=smallsize, left=False, right=True, labelleft=False, labelright=True)
axins.yaxis.set_label_position("right")
axins.errorbar(rho_coll_exp,y_Nhu, yerr=y_std, fmt='o', capsize=2.)
axins.plot(rho_coll_exp,linfit_exp_line,'--k',label='Linear Fit')
axins.text(0.02,0.45,r'$y=mx+b$'+'\n'+'m={:.2f}'.format(linfit_exp[0])+'\n'+'b={:.2f}'.format(linfit_exp[1])+'\n'+'$R^2=${:.3f}'.format(lin_exp_r2),transform = axins.transAxes,color='k', fontsize = 12);

if full_range:
    fig.savefig('copoly5_FigSI_fullrange.png',dpi=300, bbox_inches='tight')
else:
    fig.savefig('copoly5_FigSI.png',dpi=300, bbox_inches='tight')
```
